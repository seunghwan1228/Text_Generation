## additinal requirment, after the predict, change the softmax weight  which generate more create sentence

```
print(second_pred)
print()
print('argmax')
second_pred_label = np.argmax(second_pred)

print(second_pred_label)


print(int_to_word[second_pred_label])

```

```
## re-weight softmax, to create variety sentence

second_pred

temperature = 0.6
log_pred = np.log(second_pred) / temperature
log_pred

exp_pred = np.exp(log_pred)
exp_pred

scale_exp_pred = exp_pred / np.sum(exp_pred)
scale_exp_pred

np.sum(scale_exp_pred)


pred_proba = np.random.multinomial(1, scale_exp_pred[0], 1)
pred_proba

proba_argmax = np.argmax(pred_proba)
proba_argmax


print(int_to_word[proba_argmax])
```

```
# wrap as function

def re_weight_sample(pred, temperature=0.6):
  pred = pred.astype('float64')
  log_pred = np.log(pred) / temperature
  exp_pred = np.exp(log_pred)
  scale_exp_pred = exp_pred / np.sum(exp_pred)
  pred_proba = np.random.multinomial(1, scale_exp_pred[0], 1)
  proba_argmax = np.argmax(pred_proba)
  return proba_argmax, int_to_word[proba_argmax]
```

```
re_weight_sample(second_pred, 0.5)
```

## create loop

- purpose : create a 4000 character by epochs 

1) train model by 1 epoch

2) sample the seed sentence

3) create a sentence for 4000


```
import sys
```

```
model = models.Sequential()
model.add(layers.LSTM(128, input_shape=(100, 45)))
model.add(layers.Dense(45, activation='softmax'))
# model.summary()
model.compile(loss='categorical_crossentropy', optimizer='adam')



for epoch in range(1, 21):  # set epoch
  model.fit(x_cat, y_cat, batch_size=128, epochs=1)
  sentence_seed = np.random.randint(0, len(path)-100)
  
  start_index = x[sentence_seed]
  
  start_index_cat = to_categorical(start_index, num_classes=len(set(path)))
  start_index_cat = start_index_cat.reshape(1, 100, 45)
  
  sentence_seed_text = []
  for i in start_index:

    original_sentence = int_to_word[i]
    sentence_seed_text.append(original_sentence)
    full_sentence_seed = ''.join(sentence_seed_text)
  print('---------------')
  print()
  print('original_sentence : ','\n', ''.join(sentence_seed_text))
  print()
  print('---------------')
  print()
  
  for i in range(1, 1000):                    # generate 4000 character   ex) original sentence + char_1 + char_2 + char_3 ... char_4000 
    start_index_cat = to_categorical(start_index, num_classes=45).reshape(1, 100, 45)
    preds = model.predict(start_index_cat)
    weighted_softmax, weighted_character = re_weight_sample(preds, temperature=0.6)
    
    next_char = int_to_word[weighted_softmax]
    
    sentence_seed_text.append(next_char)
    
    
    start_index = np.append(start_index, np.array(weighted_softmax))
    start_index = start_index[1:]
    
    sys.stdout.write(weighted_character)
#     print('----------------')
  print()
  print('---------------')
  print()
  print(''.join(sentence_seed_text))
  print()
  print('---------------')
  print()
    
```

```
for epoch in range(1, 50):  # set epoch
  model.fit(x_cat, y_cat, batch_size=128, epochs=1)
  sentence_seed = np.random.randint(0, len(path)-100)
  
  start_index = x[sentence_seed]
  
  start_index_cat = to_categorical(start_index, num_classes=len(set(path)))
  start_index_cat = start_index_cat.reshape(1, 100, 45)
  
  sentence_seed_text = []
  for i in start_index:

    original_sentence = int_to_word[i]
    sentence_seed_text.append(original_sentence)
    full_sentence_seed = ''.join(sentence_seed_text)
  print('---------------')
  print('epoch : ', epoch)
  print()
  print('original_sentence : ')
  print()
  print(''.join(sentence_seed_text))
  print()
  print('---------------')
  print()
  
  for i in range(1, 1500):                    # generate 4000 character   ex) original sentence + char_1 + char_2 + char_3 ... char_4000 
    start_index_cat = to_categorical(start_index, num_classes=45).reshape(1, 100, 45)
    preds = model.predict(start_index_cat)
    weighted_softmax, weighted_character = re_weight_sample(preds, temperature=0.5)
    
    next_char = int_to_word[weighted_softmax]
    
    sentence_seed_text.append(next_char)
    
    
    start_index = np.append(start_index, np.array(weighted_softmax))
    start_index = start_index[1:]
    
    sys.stdout.write(weighted_character)
#     print('----------------')
  print()
  print('---------------')
  print()
  print(''.join(sentence_seed_text))
  print()
  print('---------------')
  print()
```
