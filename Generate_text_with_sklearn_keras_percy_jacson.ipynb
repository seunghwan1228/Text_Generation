# create model

```
model = models.Sequential()
model.add(layers.LSTM(128, input_shape=(500, 59)))
model.add(layers.Dense(59, activation='softmax'))
model.summary()
model.compile(loss='categorical_crossentropy', optimizer='adam')
```

```
def re_weight(y_pred, temperature=0.5):
  y_pred = np.asarray(y_pred, dtype='float64')
  log_y_pred = np.log(y_pred) / temperature
  exp_y_pred = np.exp(log_y_pred)
  norm_y_pred = exp_y_pred / np.sum(exp_y_pred)
  proba_y = np.random.multinomial(1, norm_y_pred[0], 1)
  result_index = np.argmax(proba_y)
  return result_index
```
